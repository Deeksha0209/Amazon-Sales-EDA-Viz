**#Description of Dataset**

This is an amazon sales data containg data spread across Regions & Countries; Various types of items, sales channels. It also covers Orders info such as Order Priority , Order Date, Order ID etc. This dataset also covers units sold, total profit and revenue generated et al.

# **#Let's perform Data Cleaning on df!**

```python
#checking null values. Great, there are no na values in df!
df.isnull().sum()
```

```python
#checking and dropping duplicates from df!
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
#Converting Order Date & Ship Date to correct data type & creating new columns for better analysis!
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order_Weekday'] = df['Order Date'].dt.day_name()
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Ship_Weekday'] = df['Ship Date'].dt.day_name()
```

```python
#Lets take a revised look at the data types!
df.dtypes
```

```python
#Rounding off numerical colms for better results
df['Unit Price']=df['Unit Price'].round()
df['Unit Cost']=df['Unit Cost'].round()
df['Total Cost']=df['Total Cost'].round()
df['Total Revenue']=df['Total Revenue'].round()
df['Total Profit']=df['Total Profit'].round()
```

```python
#Lets take a revised look at the shape of our df!
df.shape
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%m/%d/%Y")
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format="%m/%d/%Y")


# Let's extract weekdays for 'Order Date' i.e the co-oresponding day to the date value
df['Order_WeekdayNo'] = df['Order Date'].dt.weekday

# Let's extract weekdays for 'Ship Date' i.e the co-oresponding day to the date value
df['Ship_WeekdayNo'] = df['Ship Date'].dt.weekday
```

```python
sns.set(style="darkgrid")

# Melt the DataFrame for better FacetGrid usage to incorporate both columns
df_histograms = pd.melt(df[['Order_WeekdayNo', 'Ship_WeekdayNo']], var_name='variable', value_name='value')

# Creating a FacetGrid with two subplots s.t our histograms appear side by side!
h = sns.FacetGrid(df_histograms, col="variable", col_wrap=2, height=4)
h.map(sns.histplot, "value", bins=7)
h.set_axis_labels("Weekday", "Frequency Count")
h.set_titles("Histogram of {col_name}")
plt.show()
```

```python
#Creating barplot for bivariate analysis of Units Sold and Order Priority
plt.figure(figsize=(14,7))
sns.barplot(x=df['Order Priority'], y=df['Units Sold'], order=df['Order Priority'], palette="deep")
plt.ylabel('No of Units Sold', fontsize=12)
plt.xlabel('Types of Order Priority', fontsize=12)
plt.title('Units Sold wrt Order Priority', fontsize=14)
plt.show()
```

```python
#Creating scatterplot viz of Unit Price wrt Cost
plt.figure(figsize=(10,6))
plt.scatter(df['Unit Price'], df['Total Cost'], color='orange')
plt.xlabel('Unit Price', fontsize=12)
plt.ylabel('Total Cost', fontsize=12)
plt.title('Scatter Plot of Unit Price wrt Cost', fontsize=14)
plt.show()
```

```python
#Creating lineplot for analysis of Total Revenue wrt Units Sold
plt.figure(figsize=(10, 6))
sns.lineplot(x='Total Revenue', y='Units Sold', data=df)
plt.xlabel('Total Revenue')
plt.ylabel('Total Units Sold')
plt.title('Total Revenue v/s Units Sold')
plt.grid(True)  # Add grid lines for better visualization if needed
plt.show()
```

```python
#Creating scatterplot viz of Unit Cost wrt Profit
plt.figure(figsize=(10,6))
plt.scatter(df['Unit Cost'], df['Total Profit'], color='purple')
plt.xlabel('Unit Cost', fontsize=12)
plt.ylabel('Total Profit', fontsize=12)
plt.title('Scatter Plot of Unit Cost wrt Profit', fontsize=14)
plt.show()
```

```python
#Let's further explore our data. Here i am creating a df wherein the Order_weekly & Ship_Weekly columns are equal to Mondays.
mondayData_df=df.loc[(df['Order_WeekdayNo'] == 0) & (df['Ship_WeekdayNo'] == 0)]
mondayData_df
```

```python
#Displaying a bar chart for analysis of Sales Channel v/s Total Cost data gathered on Monday!
plt.figure(figsize=(10, 6))
sns.barplot(x='Sales Channel', y='Total Cost', data=mondayData_df, palette='rocket')
plt.xlabel('Sales Channel', fontsize=12)
plt.ylabel('Total Cost', fontsize=12)
plt.title('Total Cost of Sales Channel on Mondays', fontsize=14)
plt.show()
```

```python
#Lets display a df where we fix our Order Weekday to Tuesday & Ship Weekday to Wednesday
Data_df=df.loc[(df['Order_Weekday'] == 'Tuesday') & (df['Ship_Weekday'] == 'Wednesday')]
Data_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Item Type', y='Units Sold', data=Data_df, palette='icefire')
plt.xlabel('Types of Items', fontsize=12)
plt.ylabel('Total Units Sold', fontsize=12)
plt.title('Total Units Sold for Different Items which are ordered on Tuesday & Shipped on Wednesday.', fontsize=14)
plt.show()
```

```python
#Creating df where we get total Profit by grouping df on the basis of Order Weekday
profit_df=df.groupby('Order_Weekday')['Total Profit'].mean().round().reset_index()
profit_sortdf=profit_df.sort_values(by='Total Profit', ascending=False)
profit_sortdf.head(6)
```

```python
#Lets display a pie chart viz for Total Profit wrt Order Weekday
plt.figure(figsize=(10, 8))
sns.set_palette("coolwarm")
plt.pie(profit_sortdf.head(6)['Total Profit'], labels=profit_sortdf.head(6)['Order_Weekday'], autopct='%1.0f%%', startangle=90)
plt.title('Total Profits on Order_Weekdays', fontsize=14)
plt.show()
```

```python
#Lets create a df for Total Profits wrt Ship Weekday
profit_df=df.groupby('Ship_Weekday')['Total Profit'].mean().round().reset_index()
profit_sortdf=profit_df.sort_values(by='Total Profit', ascending=False)
profit_sortdf.head(6)
```

```python
#Displaying pie plot viz for Total Profits wrt Ship Weekday
plt.figure(figsize=(10, 8))
sns.set_palette("viridis")
plt.pie(profit_sortdf.head(6)['Total Profit'], labels=profit_sortdf.head(6)['Ship_Weekday'], autopct='%1.0f%%', startangle=90)
plt.title('Total Profits on Ship_Weekdays', fontsize=14)
plt.show()
```

```python
#Displaying a horizontal bar chart viz for bivariate analysis of total Revenue generated for Countries!
plt.figure(figsize=(10, 6))
sns.barplot(x='Total Revenue', y='Country', data=Data_df, palette='magma', orient ='h')
plt.xlabel('Total  Revenue', fontsize=12)
plt.ylabel('Countries', fontsize=12)
plt.title('Total Revenue of Different Countries when orders placed on Tuesday and shipped on Wednesday.', fontsize=14)
plt.show()
```

```python
#Creating boxplots for displaying statistical measures for Unit Price wrt Order Priority
plt.figure(figsize=(10, 7))
sns.boxplot(x='Order Priority', y='Unit Price', data = df, palette='flare')
plt.xlabel('Order Priority', fontsize=12)
plt.ylabel('Unit Price', fontsize=12)
plt.title('Boxplots of Priority Data by Unit Price', fontsize=14)
plt.show()
```

```python
#Creating boxplots for displaying the analysis between Units Sold wrt Item types
plt.figure(figsize=(10, 7))
sns.boxplot(x='Item Type', y='Units Sold', data = df, palette='viridis')
plt.xlabel('Types of Items', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.title('Boxplots of Items Data by Units Sold', fontsize=14)
plt.xticks(rotation=45)  # Rotating the x-axis labels by 45 degrees
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

#Multivariate Analysis

```python
#Lets create a pairplot for our df
sns.pairplot(df, hue='Sales Channel')
```

```python
plt.figure(figsize=(12, 6))
# Using Seaborn's lineplot to plot the time series of Sales
sns.lineplot(x='Order Date', y='Units Sold', data=df, label='Order Date', color='blue')
sns.lineplot(x='Ship Date', y='Units Sold', data=df, label='Ship Date', color='yellow')
plt.xlabel('Dates', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.title('Time Series Plot of Units Sold wrt Order & Ship Dates', fontsize=14)
```

```python
#correlation matrix is a popular method for multivariate analysis of numerical columns!
numerical_columns = ['Units Sold', 'Unit Price', 'Unit Cost']
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="rocket")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
numerical_columns = ['Total Profit', 'Total Cost', 'Total Revenue']
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="flare")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
